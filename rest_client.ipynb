```python
from requests import get, post
server = 'http://localhost:8080'
```

```python
r = get(f'{server}/api/parcheggi')
if r.status_code == 200:
    print('Elenco parcheggi:')
    for p in r.json():
        print(p)
```

```python
data = {
    'data':'13/06/2025',
    'n_turisti':1000,
    'parcheggi_esclusi[]': [],
    'linee_escluse[]' : []
}
r = post(f'{server}/api/sim', json=data)
print(r.json())
```
